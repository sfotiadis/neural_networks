```python
from network2 import Network
```

```python
import mnist_loader

training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
```

```python
net = Network([784, 400, 10])
net.SGD(training_data, epochs=30, mini_batch_size=10, eta=3.0, lmbda=0.001, evaluation_data=test_data,
        monitor_evaluation_cost=True,
        monitor_evaluation_accuracy=True,
        monitor_training_cost=True,
        monitor_training_accuracy=True)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(20, 20)
wt = net.weights[0]
# use global min / max to ensure all weights are shown on the same scale
vmin, vmax = wt.min(), wt.max()
for coef, ax in zip(wt, axes.ravel()):
    ax.matshow(coef.reshape(28, 28), cmap=plt.cm.gray, vmin=.5 * vmin,
               vmax=.5 * vmax)
    ax.set_xticks(())
    ax.set_yticks(())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 4)
wt = net.weights[1]
# use global min / max to ensure all weights are shown on the same scale
vmin, vmax = wt.min(), wt.max()
for coef, ax in zip(wt, axes.ravel()):
    ax.matshow(coef.reshape(20, 20), cmap=plt.cm.gray, vmin=.5 * vmin,
               vmax=.5 * vmax)
    ax.set_xticks(())
    ax.set_yticks(())
```

```python
templates = []
wt = net.weights[0]
for coef in wt:
    templates.append(coef.reshape(28, 28))
```

```python
wt2 = net.weights[1]
```

```python
import numpy as np

fig, axes = plt.subplots(3, 4)

for digit, ax in zip(range(10), axes.ravel()):
    d = np.zeros([28, 28])

    for t,w in zip(templates, wt2[digit,:]):
        d += t*w

    vmin, vmax = d.min(), d.max()

    ax.matshow(d, cmap=plt.cm.gray, vmin=.5 * vmin,
                   vmax=.5 * vmax)
    ax.set_xticks(())
    ax.set_yticks(())

plt.plot()
```
